```python
# Scatter plot belowe shows the plots between funding rounds and relationships
# Using this we can find the trend line
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['Funding rounds:'], df['relationships'])
ax.set_ylabel('Total funding')
ax.set_xlabel('Relationships')
plt.show()
```

```python
# Scatterplot of correlation between invested companies and investment rounds.
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['invested_companies'], df['investment_rounds'])
ax.set_ylabel('investment companies')
ax.set_xlabel('invested rounds')
plt.show()
```
